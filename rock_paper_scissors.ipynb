##  _GAME PLAY_

>This game is **USER** versus **COMPUTER** based, so if you want to play have it in mind that you will be playing with the computer. WISH YOU THE BEST OF LUCK. 

> **ENJOY!**

$$**COMPARISON**$$

**Rock** vs **Paper** = **Paper Wins , Rock Lose**
    
**Rock** vs **Scissors** = **Rock Wins , Scissors Lose**
        
**Scissors** vs **Paper** = **Scissors Wins , Paper Lose**

**Rock** vs **Rock** = **A Tie** {REPLAY}
        
**Scissors** vs **Scissors** = **A Tie** {REPLAY}

**Paper** vs **Paper** = **A Tie** {REPLAY}

```python
import random

print("Please provise your name \n")

name = ""
while name == "":
    name = input("Enter your name: ").capitalize()
print("Welcome To The Game ",name,"\n")

play_time = 1
print("The number of challenge or times you want to play is ", 5,"\n")

wins_user = 0
wins_computer = 0  
draws = 0




while play_time <= 5:
    
    print("Round ", play_time , "\n" )

    options = {
        "Rock":1,
        "Paper":2,
        "Scissors":3
    }
        # Computer Turn

    computer_choice = random.randint(1,3)

    for index in options:
        if options[index] == computer_choice:
            computer_option = index

    # User's turn
    print("You can chose between ROCK , PAPER and SCISSORS \n")

    user_choice = ""

    while user_choice != "Rock" and user_choice != "Paper" and user_choice != "Scissors":

        choice = input("Enter your choice (in lower case): ").capitalize()  
        user_choice = choice
    else:
        pass
    if user_choice == "Rock" or user_choice == "Paper" or user_choice == "Scissors":

        # Compare them
        compare = True

        while compare:

            if computer_option == "Paper" and user_choice == "Rock":
                print("Computer wins \n You lose ",name,"\n")
                print("Computer choose ", computer_option,"\n")
                compare = False
                wins_computer += 1
            elif computer_option == "Rock" and user_choice == "Scisssors":
                print("Computer wins \n You lose ",name,"\n")
                print("Computer choose ", computer_option,"\n")
                compare = False
                wins_computer += 1
            elif computer_option == "Scissors" and user_choice == "Paper":
                print("Computer wins \n You lose ",name,"\n")
                print("Computer choose ", computer_option,"\n")
                compare = False
                wins_computer += 1
            elif computer_option == user_choice:
                print("There's a tie")
                compare = False
                draws += 1
            else:
                print("You win \n ")
                print("Computer choose ", computer_option,"\n")
                compare = False
                wins_user += 1

    play_time += 1

print("\n")

if wins_computer > wins_user:
    print("Computer has the most wins, COMPUTER WINS")
elif wins_computer == wins_user:
    print("Equal Number of wins and one draw. NO WINNER")
else:
    print(f"{name} has the most numbers of wins, {name.upper()} wins")
    
    
```
